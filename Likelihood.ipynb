```python
from marcia import Sampler
import os
%matplotlib inline
```

```python
chaindir = 'Chains'
os.makedirs(chaindir, exist_ok=True)
```

```python
lcdm = Sampler('LCDM',['H0','Omega_m'],['Pantheon_plus'],[70,0.25],sampler_file='Chains/Pantheon_LCDM.h5',max_n=5000)
s = lcdm.sampler()
lcdm.corner_plot()
```

```python
wCDM = Sampler('wCDM',['H0','Omega_m','w0'],['Pantheon_plus'],[70,0.25,-1],sampler_file='Chains/Pantheon_wCDM.h5',max_n=5000)
s = wCDM.sampler()
wCDM.corner_plot()
```

```python
cpl =  Sampler('CPL',['H0','Omega_m','w0','wa'],['Pantheon_plus'],[70,0.25,-1,0],sampler_file='Chains/Pantheon_CPL.h5',max_n=6000,converge=True)
s = cpl.sampler()
cpl.corner_plot()
```
